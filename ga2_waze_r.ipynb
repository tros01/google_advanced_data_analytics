```r
df <- read_csv(
    file.path(r"(data/waze_dataset.csv)"),
    col_names = TRUE,
    col_types = NULL
)
```

```r
# Check df dimensions
expected_shape <- c(14999, 13)
sapply(seq(length(expected_shape)), \(i) expected_shape[i] == dim(df)[i])
```

```r
# View the top rows
df |> head(n=10)
```

```r
# Inspect the structure
df |> str()
```

```r
# Isolate rows with null values
df_filtered_null <- df |>
    filter(if_any(everything(), is.na))

df_filtered_null |> head(n=10)
```

```r
# Display summary stats of rows with null values
df_filtered_null |>
    summary()
```

```r
# Isolate rows without null values
df_filtered_nonnull <- df |>
    filter(!if_any(everything(), is.na))

df_filtered_nonnull |> head(n=10)
```

```r
# Display summary stats of rows without null values
df_filtered_nonnull |> 
    summary()
```

```r
# Get count of null values by device
df_null <- df |>
    group_by(device) |>
    summarise(
        across(
                .cols = everything(),
                .fns = ~sum(is.na(.x))
            )
        ) |>
    ungroup()
df_null
```

```r
# Calculate % of iPhone nulls and Android nulls
df_null |>
    rowwise() |>
    mutate(
        null_sum = sum(c_across(where(is.numeric)), na.rm = TRUE)
        # null_sum = rowSums(across(where(is.numeric)), na.rm = TRUE) # Without rowwise()
    ) |>
    ungroup() |>
    select(device, null_sum) |>
    mutate(null_sum = null_sum / sum(null_sum))
```

```r
# Filter out NaNs to obtain a balanced df
df_balanced <- df |>
    drop_na() |>
    print()
```

```r
# Calculate % of iPhone users and Android users in full dataset
df |>
    group_by(device) |>
    tally() |>
    mutate(n = n / nrow(df))
```

```r
# Calculate counts of churned vs. retained
label_count <- df |>
    group_by(label) |>
    tally() |>
    ungroup()
label_count |> print()
```

```r
# Percentages of churned and retained, excluding NAs
label_pc <- label_count |>
    filter(!is.na(label)) |>
    mutate(pc = n / sum(n)) |>
    print()
```

```r
# Calculate median values of all columns for churned and retained users
df |>
    group_by(label) |>
    summarise(
        across(
            .cols = where(is.numeric),
            .fns = ~ median(.x)
        ),
        .groups = "drop"
    )
```

```r
# Add a column to df called `km_per_drive`
df <- df |>
    mutate(km_per_drive = driven_km_drives / drives)
df |>
    select(label, ID, km_per_drive) |>
    head(n=10)
```

```r
# Group by `label`, calculate the median, and isolate for km per drive
df |>
    group_by(label) |>
    summarise(
        count = n(),
        median_km_per_drive = median(km_per_drive, na.rm = TRUE),
        .groups = "drop"
    )
```

```r
# Add a column to df called `km_per_driving_day`
# Place the new col just after km_per_drive
df <- df |>
    add_column(km_per_driving_day = df$driven_km_drives / df$driving_days, .after = "km_per_drive")
df |> select(label, ID, km_per_driving_day) |> head(n=10)
```

```r
# Group by `label`, calculate the median, and isolate for km per driving day
stats_df <- df |>
    mutate(
        across(
            # Clean up Infs and 0s
            .cols = c("km_per_drive", "km_per_driving_day"),
            .fns = ~ ifelse(is.infinite(.x) | .x == 0, NA, .x)
        )
    ) |>
    group_by(label) |>
    summarise(
        across(
            .cols = c("km_per_drive", "km_per_driving_day"),
            .fns = list(
                mean = ~ mean(.x, na.rm = TRUE),
                median = ~ median(.x, na.rm = TRUE),
                sd = ~ sd(.x, na.rm = TRUE)
            ),
            .names = "{.col}_{.fn}"
            )
        ) |>
    ungroup() 
stats_df |> head(n=10)
```

```r
# Add a column to df called `drives_per_driving_day`
df <- df |>
    mutate(
        drives_per_driving_day = drives / driving_days
    ) |>
    relocate(drives_per_driving_day, .after = "km_per_driving_day") # Or get index with which()

df |> select(drives_per_driving_day) |> tail(n=10)
```

```r
# Group by `label`, calculate the median, and isolate for drives per driving day
df |>
    group_by(label) |>
    summarise(
        drives_per_driving_day_median = median(drives_per_driving_day, na.rm = TRUE), 
        .groups = "drop"
    ) |>
    ungroup()
```

```r
# For each label, calculate the number of Android users and iPhone users
df |>
    group_by(label, device) |>
    tally() |> # Or count()
    ungroup()
```
