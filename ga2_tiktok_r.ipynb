```r
df <- read_csv(
    file.path(r"(data/tiktok_dataset.csv)"),
    col_names = TRUE,
    col_types = NULL
)
```

```r
# Check df dimensions
expected_shape <- c(19382, 12)
sapply(seq(length(expected_shape)), \(i) expected_shape[i] == dim(df)[i])
```

```r
# Inspect the first 10 rows
df |> head(n=10)
```

```r
# Examine the structure
df |> str()
```

```r
# Clean the cols
df <- df |>
    janitor::clean_names()
df |> colnames()
```

```r
# Fix data types
df <- df |>
    mutate(
        across(
            .cols = ends_with("count"),
            .fns = ~ as.numeric(.x)
        )
    )
```

```r
# Inspect NAs
df |>
    select(ends_with("count")) |>
    summarise(
        across(
            .cols = everything(),
            .fns = ~sum(is.na(.x))
        )
    )
```

```r
# Inspect the summary stats
df |> summary()
```

```r
# What are the different values for claim status and how many of each are in the data?
df |>
    count(claim_status)
```

```r
# What is the average view count of videos with each status?
df |>
    group_by(claim_status) |>
    summarise(avg_status_count = mean(video_view_count)) |>
    ungroup()
```

```r
# What is the average view count of videos with "opinion" status?
df |>
    group_by(claim_status) |>
    summarise(avg_status_count = mean(video_view_count)) |>
    filter(claim_status  == "opinion") |>
    ungroup()
```

```r
# Get counts for each group combination of claim status and author ban status
df |>
    group_by(claim_status) |>
    count(author_ban_status) |>
    ungroup()
```

```r
# What's the median video share count of each author ban status?
df |>
    group_by(author_ban_status) |>
    summarise(
        median_share_count = median(video_share_count, na.rm = TRUE)
    ) |>
    ungroup()
```

```r
# Count, mean median by ban status
df |>
    group_by(author_ban_status) |>
    summarise(
        across(
            .cols = c(video_view_count, video_like_count, video_share_count),
            .fns = list(
                count = ~n(),
                mean = ~mean(.x, na.rm = TRUE),
                median = ~median(.x, na.rm = TRUE)
            )
        )
    ) |>
    ungroup()
```

```r
# Create a likes/comments/shares_per_view column
ratio_cols <- c("likes_per_view", "comments_per_view", "shares_per_view")

for (col in ratio_cols) {
    key_term <- str_extract(col, "^[a-z]+(?=s)")
    source_col <- paste0("video_", key_term, "_count")

    df[col] <- df[source_col] / df["video_view_count"]
}

df[ratio_cols] |> head(n=15)
```

```r
# Summary stats for the new cols
df |>
    group_by(claim_status, author_ban_status) |>
    summarise(
        across(
            .cols = all_of(ratio_cols),
            .fns = list(
                count = ~n(),
                mean = ~mean(.x, na.rm = TRUE),
                median = ~median(.x, na.rm = TRUE)
            )
        )
    ) |>
    ungroup()
```
